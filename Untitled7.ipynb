```python
import pygame
import time
from pygame.locals import *
import random

class playerPlane(object):
    def __init__(self,screen):
        self.playerName = 'plane'
        playerImage = 'plane.png'
        self.Image = pygame.image.load(playerImage).convert()
        self.x = 5
        self.y = 520
        
        self.bullet = []
        self.screen = screen
        
    def draw(self):
        self.screen.blit(self.Image,(self.x,self.y))
        for temp in self.bullet:
            temp.draw2()
            
    def playerRange(self):
        if self.x < 0:
            self.x = 0
        elif self.x > 1000:
            self.x = 1000
        elif self.y < 0:
            self.y = 0
        elif self.y > 520:
            self.y = 520
            
    def keyHandle(self,keyValue):
        global G_Bullet
        if keyValue == 'left':
            self.x -= 20
            self.playerRange()
        elif keyValue == 'right':
            self.x += 20
            self.playerRange()            
        elif keyValue == 'up':
            self.y -= 20  
            self.playerRange()            
        elif keyValue == 'down':
            self.y += 20            
            self.playerRange()
        elif keyValue == 'space':
            print('发射导弹')
            self.bullet.append(Bullet(self.screen,self.x,self.y))

class Bullet(object):
    def __init__(self,screen,x,y):
        bulletImage = 'cutecute.png'
        self.ziDan = pygame.image.load(bulletImage).convert()
        self.screen = screen 
        self.x = x 
        self.y = y 
        
    def draw2(self):
        self.y -= 5
        self.screen.blit(self.ziDan,(self.x,self.y))
        
class EnemyPlane(object):
    def __init__(self,screen):
        self.name = 'enemy'
        enemyPlaneImage = 'badbad.png'
        self.image = pygame.image.load(enemyPlaneImage).convert()
        self.x = 0 
        self.y = 0 
        self.enemyScreen =screen
        self.direction = 'right'
        self.enemyBul = []
        self.enemyplane1 = [0,]
        self.enemyplane2 = [0,]
        self.enemybullet = [0,]
    
    def moveEnemy(self):
        if self.x >= 1000:
            self.direction = 'left'
        elif self.x < 0 :
            self.direction = 'right'
        if self.direction == 'right':
            self.x += 10
        elif self.direction == 'left':
            self.x -= 10
        self.y +=1
        randomNum1 = random.randint(0,10)
        randomNum2 = random.randint(0,1000)
        if randomNum2 % 5 == 0 and len(self.enemyplane1)<100:
            self.enemyplane1.append(randomNum2+self.x)    
            
        elif randomNum2 % 5 == 0 and len(self.enemyplane1)>100:
            self.enemyplane1[100]=randomNum2+self.x
            for i in range(0,100):
                self.enemyplane1[i]=self.enemyplane1[i+1]
            self.enemyplane1.pop()
            if randomNum1 in [1,2,3,4,5,6,8,10] :
                self.enemyBul.append(enemyBullet(screen,randomNum2+self.x,self.y))  
      
        
   
                
    def draw3(self):
        self.moveEnemy()
        
        for n in self.enemyplane1: 
            if self.x+n>1000:
                randomNum4 = random.randint(-5,5)
                self.enemyScreen.blit(self.image,(self.x+n-1000,0))         
            elif self.x+n<= 1000:
                randomNum4 = random.randint(-5,5)
                self.enemyScreen.blit(self.image,(self.x,self.y+randomNum4))
                

            
        for temp in self.enemyBul:
            temp.draw4()
            
class enemyBullet(object):
    def __init__(self,screen,x,y):
        enemyBul = 'shit.png'
        self.enemyBulImage = pygame.image.load(enemyBul).convert()
        self.x = x - 5
        self.y = y + 30
        self.eneScreen = screen
    def draw4(self):
        self.y += 10
        self.eneScreen.blit(self.enemyBulImage,(self.x,self.y))
        
def checkDead():
    cuteRect = pygame.Rect(Pipeline.wallx,0+Pipeline.x,
                         Pipeline.pineUp.get_width() - 10,
                         Pipeline.pineUp.get_height())
    downRect = pygame.Rect(Pipeline.wallx,400-Pipeline.y,
                         Pipeline.pineDown.get_width() - 10,
                         Pipeline.pineDown.get_height())
    
    if upRect.colliderect(Bird.birdRect) or downRect.colliderect(Bird.birdRect):
        Bird.dead = True
    if not 0 < Bird.birdRect[1] < height:
        Bird.dead = True
        return True
    else:
        return False
            
if __name__ == '__main__':
    screen = pygame.display.set_mode((1000,562),0,32)
    bgImageFile = 'board.png'
    backGround = pygame.image.load(bgImageFile).convert()
    player = playerPlane(screen)
    enemyPlane = EnemyPlane(screen)
    while True:
        screen.blit(backGround,(0,0))
        for event in pygame.event.get():
            if event.type == QUIT:

                print('退出')
                quit()
            elif event.type == KEYDOWN:
                if event.key == K_a or event.key == K_LEFT:
                    print('left')
                    player.keyHandle('left')
                elif event.key == K_d or event.key == K_RIGHT:
                    print('right')
                    player.keyHandle('right')
                elif event.key == K_w or event.key == K_UP:
                    print('up')
                    player.keyHandle('up')
                elif event.key == K_s or event.key == K_DOWN:
                    print('down')
                    player.keyHandle('down')
                elif event.key == K_SPACE:
                    print('space')
                    player.keyHandle('space')
   
                    
        player.draw()
        enemyPlane.draw3()
        pygame.display.update()
        time.sleep(0.1)
    
    
    
    
    
    
    
    
    

            
            
            
            
            
            
            
            
```
